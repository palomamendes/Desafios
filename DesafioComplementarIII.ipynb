```python
import pandas as pd
```

```python
# Exercicio I

game = pd.read_excel("best-selling game consoles.xlsx",
                      sheet_name="consoles")
game = game.groupby('Company')[['Company']].count()
game
```

```python
# Exercicio II
game.columns = ['Qtd Games']
game = (game
              .sort_values('Qtd Games',ascending=False)
              .reset_index())
game
```

```python
#Execicio III

company = pd.read_csv("company.txt")
company
```

```python
#Execicio III

game_merge = game.merge(company,on='Company',how='left')
game_merge
```

```python
#Execicio IV

market_sum =  game_merge.sum()
market_med = round(market_sum['Market Value']/market_sum['Qtd Games'], 2)
print("Market Value : ", market_med)
```
